```python
# imports
import pandas as pd
import numpy as np
import os
from transformers import pipeline
```

```python
csv_path = r"{}\updated_news_data.csv".format(current_dir)

if test:
    print('csv_path: ' + str(csv_path) + '\n')

# convert csv to pandas dataframe
news_data = pd.read_csv(csv_path)

if test:
    print('len(news_data)    :' + str(len(news_data))    + '\n')
    print('news_data.iloc[0] :' + str(news_data.iloc[0]) + '\n')
    print('news_data.head()  :' + str(news_data.head())  + '\n')
```

```python
#  import 
from textblob import TextBlob
```

```python
# function for sentiment using textblob
def get_sentiment(text):
    if isinstance(text, str):
        analysis = TextBlob(text)
        return analysis.sentiment.polarity
    else:
        return None

# loop from 1 and 25
for i in range(1, 26):
    news_data[f'Sen_Top{i}'] = news_data[f'Top{i}'].apply(get_sentiment)

# loop through first 20 and print out the sentiment to test
for i in range(1, 26):
    print(news_data.iloc[0][f'Top{i}'])
    print(news_data.iloc[0][f'Sen_Top{i}'])
```
